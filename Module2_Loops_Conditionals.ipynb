## 4.1
Choose a number between 1 and 10 and assign it to the variable `secret`. Then choose another number between 1 and 10 and assign it to the variable `guess`. Write conditional tests to print:
- `too low` if `guess` is less than `secret`
- `too high` if `guess` is greater than `secret`
- `just right` if `guess` is equal to `secret`

```python
secret = 7
guess = 5

if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

Assign `True` or `False` to the variables `small` and `green`. Then use if/else statements to print which item matches those choices:

## 6.2

This exercise uses a while loop to compare the variable `number` to `guess_me`. The loop prints **too low** while the number is less than `guess_me`, prints **found it!** when the values match, and stops. If the number becomes greater than `guess_me`, it prints **oops** and exits the loop.

```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
    number += 1
```

## 6.3

This exercise uses a for loop with `range(10)` to compare each value of `number` to `guess_me`. The program prints **too low** when the number is smaller, **found it!** when the number matches, and **oops** if the number becomes greater than `guess_me`.

```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
```
